```python
import pandas as pd
import numpy as np
from sklearn.svm import SVC
from sklearn import metrics
from sklearn.model_selection import train_test_split
```

```python
df = pd.read_csv(r"C:\Users\alokk\Desktop\csv\pima-diabetes.csv")
df.head()
```

```python
def fitting(x,y,C,gamma):
    
    x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state = 0)
    iters = ['rbf','poly','sigmoid']
    
    for i in iters:
        
        classifier = SVC(kernel = i, probability = True, C = C, gamma = gamma)
        classifier.fit(x_train,y_train)
        
        y_pred_train = classifier.predict(x_train) #train
        y_pred_test = classifier.predict(x_test)  #test
        
        print("----------Evaluation on Test Data------------",i)
        score_test = classifier.score(x_test,y_test)
        print("Accuracy Score:", score_test)
        print("Classification Report:\n",metrics.classification_report(y_test,y_pred_test))
        
        print("----------Evaluation on Training Data---------",i)
        score_train = classifier.score(x_train,y_train)
        print("Accuracy Score:",score_train)
        print("Classification Report:\n",metrics.classification_report(y_train,y_pred_train))
        print("-----------------------------------------------")
```

```python
x = df.drop(['Outcome'],axis = 1).values
y = df['Outcome'].values

fitting(x,y,1,'scale')
```
